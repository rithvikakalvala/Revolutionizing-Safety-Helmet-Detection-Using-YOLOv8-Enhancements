```python
from ultralytics import YOLO
import cv2
import os
from pathlib import Path
```

```python
# Set dataset directory path
dataset_dir = "C:/Users/verti/Desktop/HELMENT_NEW/abcd.v3i.yolov8"

# Example structure
train_images = os.path.join(dataset_dir, "train/images")
val_images = os.path.join(dataset_dir, "valid/images")
train_labels = os.path.join(dataset_dir, "train/labels")
val_labels = os.path.join(dataset_dir, "valid/labels")
```

```python
# Create the YAML file for the dataset
yaml_content = """
path: C:/Users/verti/Desktop/HELMENT_NEW/abcd.v3i.yolov8
train: C:/Users/verti/Desktop/HELMENT_NEW/abcd.v3i.yolov8/train
val: C:/Users/verti/Desktop/HELMENT_NEW/abcd.v3i.yolov8/valid

nc: 2  # Number of classes
names: ['helmet', 'with out helmet']  # Replace with your class names
"""

# Save the YAML file
with open("dataset.yaml", "w") as f:
    f.write(yaml_content)
```

```python
# Initialize YOLOv8 model (e.g., yolov8n for Nano version)
model = YOLO('yolov8n.pt')  # Or load a pretrained model like 'yolov8n.pt'
```

```python
# Train the YOLOv8 model
model.train(data='dataset.yaml', epochs=10, imgsz=640, batch=16)
```

```python
# Evaluate the trained model on validation data
metrics = model.val()
print(metrics)
```

```python
#Image code


from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO(r'C:/Users/verti/Desktop/HELMENT_NEW/abcd.v3i.yolov8/runs/detect/train28/weights/best.pt')

# Input and output image paths
input_image_path = r"C:/Users/verti/Desktop/HELMENT_NEW/abcd.v3i.yolov8/test/images/hard_hat_workers149_png.rf.6113d10a19cd7acae9b3d5a593f10e99.jpg"
output_image_path = r"C:/Users/verti/Desktop/HELMENT_NEW/abcd.v3i.yolov8/output1.jpg"

# Load the image
image = cv2.imread(input_image_path)

# Perform detection
results = model.predict(image)

# Annotate detections
for result in results[0].boxes.data.tolist():
    x1, y1, x2, y2, conf, cls = result
    x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))  # Bounding box coordinates
    label = f"{model.names[int(cls)]} {conf:.2f}"  # Label with confidence

    # Draw bounding box and label
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Save the output image
cv2.imwrite(output_image_path, image)


```
